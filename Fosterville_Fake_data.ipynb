```python
import pandas as pd
import random as rd
```

```python
# Create Column date
months = ['2019-09', '2019-10', '2019-11', '2019-12']

period = pd.Period('2019-08', freq='M')
dataset = pd.Series(pd.DatetimeIndex(start=period.start_time, end=period.end_time, freq='D')).to_frame('date')

for month in months:
    period = pd.Period(month, freq='M')
    dataset = pd.concat([dataset, 
                     pd.Series(pd.DatetimeIndex(start=period.start_time, end=period.end_time, freq='D')).to_frame('date')],
                   ignore_index=True)
```

```python
machine_type_list = ["Bogger", "Truck", "Jumbo", "Solo", "Charge"]
```

```python
dataset['Machine'] = 0

# Create Column Machine Type
for i in range(len(dataset)): 
    ##dataset['Machine'][i] = (rd.choice(machine_type_list))
    # Use .loc to prevent create copy
    dataset['Machine'].loc[i] = (rd.choice(machine_type_list))
```

```python
# This function will generate hour from 0 to 12 but more bias toward 12 more.
def generate_time_bias():
    rdNumber = rd.randint(0, 25)
    if rdNumber > 12:
        rdNumber = 12
    return rdNumber
```

```python
# Add 2 new columns Day-Up and Night-Up
dataset['Day-Up'] = 0
dataset['Night-Up'] = 0

# Create Columns Day-Up and Night-Up
for i in range(len(dataset)): 
    dataset['Day-Up'].loc[i] = generate_time_bias()
    dataset['Night-Up'].loc[i] = generate_time_bias()
```

```python
# Add 2 new columns Day-Down and Night-Down
dataset['Day-Down'] = 0
dataset['Night-Down'] = 0

# Create Columns Day-Up and Night-Up
for i in range(len(dataset)): 
    dataset['Day-Down'].loc[i] = 12 - dataset['Day-Up'].loc[i]
    dataset['Night-Down'].loc[i] = 12 - dataset['Night-Up'].loc[i]
```

```python
# Add Percentage-Down and Percentage-Up
dataset['Percentage-Up'] = 0
dataset['Percentage-Down'] = 0
dataset['Downtime Hours'] = 0

# Create Columns Percentage-Up and Percentage-Down
for i in range(len(dataset)): 
    dataset['Percentage-Up'].loc[i] = round((dataset['Night-Up'].loc[i] + dataset['Day-Up'].loc[i]) / 24, 2)
    dataset['Percentage-Down'].loc[i] = round((dataset['Night-Down'].loc[i] + dataset['Day-Down'].loc[i]) / 24, 2)
    dataset['Downtime Hours'].loc[i] = (dataset['Night-Down'].loc[i] + dataset['Day-Down'].loc[i])
```

#### Add Column Downtime and Downtime Reason

```python
downTime = ["Accident Damage", "Breakdown", "Inspection", "Operator Error", "Planned"]

downTimeReason = ["Awaiting Approval", "Awaiting Labour", "Awaiting Other Resources", "Awaiting Parts", 
            "Repair Inspection", "Safety Inspection", "Under Repair", "Weather Delay"]
```

```python
# Add Columns Downtime and Downtime Reason
dataset['Downtime'] = 0
dataset['Downtime Reason'] = 0

# Create Columns Downtime and Downtime Reason
for i in range(len(dataset)): 
    dataset['Downtime'].loc[i] = (rd.choice(downTime))
    dataset['Downtime Reason'].loc[i] = (rd.choice(downTimeReason))
```
